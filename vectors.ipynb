```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
plt.quiver(0,0,3,4)
plt.show()
```

```python
plt.quiver(0,0,3,4, scale=1, scale_units='xy', angles='xy', color = 'red')
plt.quiver(0,0,-3,4, scale=1, scale_units='xy', angles='xy', color = 'blue')

```

```python
def plot_vectors(vect):
    i = 0
    colors = ['r', 'b', 'g' , 'y']
    for vec in vect:
        plt.quiver(vec[0],vec[1],vec[2],vec[3], scale=1, scale_units='xy', angles='xy', color = colors[i % len(colors)])
        i += 1
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    plt.show()
```

```python
plot_vectors([(0,0,3,4), (0,0,-3,4), (0,0,-3,-2), (0,0,4,-1)])
```

```python
vecs = [np.array([0,0,3,4]), np.array([0,0,-3,4]), np.array([0,0,-3,-2]), np.array([0,0,4,-1])]
```

```python
plot_vectors([vecs[0], vecs[3]])
```

```python
vecs[0] + vecs[1]
```

```python
plot_vectors([vecs[0], vecs[2], vecs[0] + vecs[2]])
```

```python
plot_vectors([vecs[0], vecs[3], vecs[0] + vecs[3]])
```

```python
plot_vectors([vecs[0], vecs[0], vecs[0] + vecs[0]])
```
